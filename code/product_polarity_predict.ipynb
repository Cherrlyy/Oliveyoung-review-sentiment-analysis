```python
import pandas as pd
```

```python
review = pd.read_csv('review_final_preprocessed.csv', encoding='UTF-8')
type_polarity = pd.read_csv('review_type_polarity.csv', encoding='UTF-8')
```

```python
review
```

```python
type_polarity
```

```python
product = review.drop_duplicates(['name'])
product
```

```python
review_polarity = review.join(type_polarity)
review_polarity
```

```python
review_polarity = review.join(type_polarity)
review_polarity_sum = review_polarity.groupby('name').sum()
review_polarity_cnt = pd.DataFrame(review_polarity.groupby('name')['content'].count())
review_polarity_cnt
```

```python
review_polarity = review_polarity_sum.join(review_polarity_cnt, on='name')
review_polarity
```

```python
review_polarity_ratio = pd.DataFrame()
review_polarity_ratio['soothing'] = review_polarity['soothing']/review_polarity['content']
review_polarity_ratio['moisturizing'] = review_polarity['moisturizing']/review_polarity['content']
review_polarity_ratio['skin'] = review_polarity['skin']/review_polarity['content']

review_polarity_ratio
```

```python
product_polarity = product.join(review_polarity_ratio, on='name')
product_polarity
# ratio : 리뷰 1개당 평균 긍정도
```

```python
product_polarity.to_csv("product_polarity.csv", index=False)
```
